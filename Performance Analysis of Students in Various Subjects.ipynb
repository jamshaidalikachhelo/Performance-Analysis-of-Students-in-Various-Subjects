```python
def find_percentage(obtained_marks):
    percentage = (obtained_marks / 100) * 100
    if percentage >= 90:
        return 'A+'
    elif percentage >= 80:
        return 'A'
    elif percentage >= 70:
        return 'B'
    elif percentage >= 60:
        return 'C'
    elif percentage <= 50:
        return 'Fail'

df['Math_Percentage'] = df['Maths Marks out of 100'].apply(find_percentage)
df['Physics_Percentage'] = df['Physics Marks out of 100'].apply(find_percentage)

print(df)
```

```python
def find_percentage(obtained_marks):
    percentage = (obtained_marks / 50) * 100
    if percentage >= 90:
        return 'A+'
    elif percentage >= 80:
        return 'A'
    elif percentage >= 70:
        return 'B'
    elif percentage >= 60:
        return 'C'
    elif percentage <= 50:
        return 'Fail'


df['Chem_Percentage'] = df['Chemistry Marks out of 50'].apply(find_percentage)
df['Arts_Percentage'] = df['Arts Project Marks out of 50'].apply(find_percentage)

```

```python
import matplotlib.pyplot as plt

def count_grades(subject_grade_column):
    return subject_grade_column.value_counts().sort_index()

fig, axs = plt.subplots(2, 2, figsize=(8, 6))

math_counts = count_grades(df['Math_Percentage'])
axs[0, 0].bar(math_counts.index, math_counts.values, color='skyblue')
axs[0, 0].set_title('Math Grade Distribution')

physics_counts = count_grades(df['Physics_Percentage'])
axs[0, 1].bar(physics_counts.index, physics_counts.values, color='lightgreen')
axs[0, 1].set_title('Physics Grade Distribution')

chem_counts = count_grades(df['Chem_Percentage'])
axs[1, 0].bar(chem_counts.index, chem_counts.values, color='salmon')
axs[1, 0].set_title('Chemistry Grade Distribution')

arts_counts = count_grades(df['Arts_Percentage'])
axs[1, 1].bar(arts_counts.index, arts_counts.values, color='orange')
axs[1, 1].set_title('Arts Grade Distribution')

plt.tight_layout()

plt.show()
```
